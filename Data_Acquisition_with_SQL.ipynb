# Loading Data

```python
# Loading the SQL extension
# ---
%load_ext sql

# Connecting to the in-memory SQLite database Google Colab
# ---
#  
%sql sqlite://
```

Loading our dataset

```python
# Loading the pandas library
# ---
#
import pandas as pd

# Step 1: Reading the dataset from the given collab folder using the pandas library.
# and storing the dataset in a variable called 'road_crashes'.
# ---
#

#Note - before exporting the data we select data for only the last 15 years using the query SELECT * FROM `traffic_accidents` WHERE `VEHICLE_YEAR`<2005

road_crashes = pd.read_csv('/content/traffic_accidents.csv') 

# Step 2: We check whether there's a table with the intended new table name.
# ---
# 
%sql DROP TABLE if EXISTS road_crashes;

# Step 3: We create a new table called 'road_crashe' within our in memory database. The persist 
# command takes the table from the 'road_crashes' variable and stores it in the database.
# ---
#
%sql PERSIST road_crashes;

# Step 4: We  select the first five records of the created road_crashes table.
# ---
#
%sql SELECT * FROM road_crashes LIMIT 5;
```

# Analysis


Display a list of 5 makes that had the highest number of crashes

```python
%sql SELECT MAKE, COUNT(MAKE) FROM road_crashes GROUP BY MAKE ORDER BY COUNT(MAKE) DESC LIMIT 5;
```

Most personal vehicles involved in crashes were traveling towards which direction?

```python
%sql SELECT TRAVEL_DIRECTION, COUNT(TRAVEL_DIRECTION) FROM road_crashes WHERE VEHICLE_USE == 'PERSONAL' GROUP BY TRAVEL_DIRECTION ORDER BY COUNT(TRAVEL_DIRECTION) DESC;
```

How many passengers were involved in personal cars were in accidents?

```python
%sql SELECT SUM(NUM_PASSENGERS) FROM road_crashes WHERE VEHICLE_USE == 'PERSONAL';
```

How many passengers were involved in accidents while traveling in personal Sport
Utility vehicles?


```python
%sql SELECT SUM(NUM_PASSENGERS) FROM road_crashes WHERE VEHICLE_TYPE == 'SPORT UTILITY VEHICLE (SUV)'; 
```

Compute the total number of Mercedez-Benz vehicles involved in accidents while
being parked

```python
%sql SELECT MAKE, COUNT(MAKE) FROM road_crashes WHERE MAKE == 'MERCEDES-BENZ' AND MANEUVER == 'PARKED'; 
```

Compute the total occupant count per vehicle count.

```python
%sql SELECT MAKE, SUM(OCCUPANT_CNT) FROM road_crashes GROUP BY MAKE ORDER BY COUNT(OCCUPANT_CNT);
```

Sort the number of passengers in descending order by travel direction taking into
account travel direction. Return travel direction and no. of passengers.


```python
%sql SELECT TRAVEL_DIRECTION, SUM(NUM_PASSENGERS) FROM road_crashes GROUP BY TRAVEL_DIRECTION ORDER BY SUM(NUM_PASSENGERS);
```

Select the top 5 vehicle models involved in the accident with no defect, order them
by make, and show their maneuver.


```python
%sql SELECT MODEL, COUNT(MODEL) FROM road_crashes WHERE VEHICLE_DEFECT =='NONE'  GROUP BY MODEL ORDER BY COUNT(MODEL) DESC LIMIT 5;
```

Which ridesharing vehicles in an accident had the most number of passengers?
Return model, and the number of passengers.


```python
%sql SELECT MODEL, MAX(NUM_PASSENGERS) FROM road_crashes;
```
